```python
import finterstellar as fs
```

```python
import os
from dotenv import load_dotenv

load_dotenv()

MY_ENV_VAR = os.getenv('OTP_VARIABLE')
```

```python
df = fs.fn_single(otp=MY_ENV_VAR, symbol='KO', window='T') # OTP: One-time-password # API Key
```

```python
# display maximum columns in pandas
import pandas as pd
pd.set_option('display.max_columns', None)
df.head()
```

```python
df['ROC'] = df['EBIT'] / ( df['Total Assets'] - df['Current Liabilities'] )
df["ROE"] = df["Net Income"] / df['Total Assets']
df["Net Cash"] = df["Cash & Equivalents"] - df["Current Liabilities"]
df["Market Cap"] = df["EPS"] * df["Shares"]
df["EV/EBITDA"] = ( df["Market Cap"] - df["Net Cash"] ) / df["EBITDA"]
```

```python
fs.draw_chart(df, left='ROE', right='Market Cap')
```

```python
fs.draw_chart(df, left='EV/EBITDA', right='Market Cap')
```

```python
# Magic formular
terms = fs.set_terms(trade_start='2011Q1', trade_end='2021Q1')
data = {}
for t in terms:
    data[t] = fs.fn_consolidated(otp=MY_ENV_VAR, term=t)
s1 = {}
s2 = {}
s = {}
signal = {}
for t in terms:
    prev_t = fs.quarters_before(terms, t, 4) # TTS: Trading term start
    data[t]["ROE"] = data[t]["Net Income"] / data[t]['Total Assets']
    data[t]["Net Cash"] = data[t]["Cash & Equivalents"] - data[t]["Current Liabilities"]
    data[t]["Market Cap"] = data[t]["Price"] * data[t]["Shares"]
    data[t]["EV/EBITDA"] = ( data[t]["Market Cap"] - data[t]["Net Cash"] ) / data[t]["EBITDA"]
    s1[t] = fs.fn_score(data[t], by='EV/EBITDA', method='relative', floor=1, cap=10, asc=False)
    s2[t] = fs.fn_score(data[t], by='ROE', method='relative', floor=0, asc=True)
    s[t] = fs.combine_score(s1[t], s2[t], n=30)
    signal[t] = list(s[t].index)
df = fs.backtest(signal=signal, data=data, m=3, cost=.001, rf_rate=.01)
```

```python
fs.draw_return(df)
```

```python
# PER + GP/A
# terms = fs.set_terms(trade_start='2011Q1', trade_end='2021Q1')
# data = {}
# for t in terms:
#     data[t] = fs.fn_consolidated(otp='OTP', term=t)
s = {}
s1 = {}
s2 = {}
signal = {}
for t in terms:
    prev_t = fs.quarters_before(terms, t, 4)
    data[t]['PER'] = data[t]['Price_M3'] / data[t]['EPS']
    data[t]['Avg Assets'] = ( data[t]['Total Assets'] + data[prev_t]['Total Assets'] ) / 2
    
    data[t]['GP/A'] = data[t]['Gross Profit'] / data[t]['Avg Assets']
    
    # price
    s1[t] = fs.fn_score(data[t], by='PER', method='relative', floor=.1, cap=2, asc=True)

    # quality
    s2[t] = fs.fn_score(data[t], by='GP/A', method='relative', floor=0, asc=False)

    # combine price factor and quality factor: pick best 30 companies
    s[t] = fs.combine_score(s1[t], s2[t], n=30)    
    signal[t] = list(s[t].index)
df = fs.backtest(signal=signal, data=data, m=3, cost=.001, rf_rate=.01)
```

```python
fs.draw_return(df) # NASDAQ companies backtest
```
