### Test Joel Greenblatt's magic formula

```python
# price factor
def factor_as_greenblatt(df):
    df["Market Cap"] = df["Price"] * df["Shares"]
    df["EV"] = df["Market Cap"] + df['Total Liabilities'] - df["Cash & Equivalents"]
    df["EV/EBIT"] = df["EV"] / df["EBIT"]
    df['ROC'] = df['EBIT'] / ( df['Total Assets'] - df['Current Liabilities'] )
    return df
```

```python
df = factor_as_greenblatt(df)
print(df.columns)
fs.draw_chart(df, left='ROC', right='Market Cap')
```

```python
from IPython.display import clear_output
def magic_formula_backtest(
    price_factor, 
    quality_factor, 
    factor_function, 
    start='2011Q1', 
    end='2021Q1',
    window_size=4,
    top_k = 30
    ):
    
    terms = fs.set_terms(trade_start=start, trade_end=end)
    data = {}
    for t in terms:
        data[t] = fs.fn_consolidated(otp=MY_ENV_VAR, term=t)
    clear_output()
    print("data succesfully imported")
    s1 = {}
    s2 = {}
    s = {}
    signal = {}
    for t in terms:
        prev_t = fs.quarters_before(terms, t, window_size) # YoY Comparison
        
        data[t] = factor_function(data[t])
        
        # score of the price factor: smallest the best
        s1[t] = fs.fn_score(data[t], by=price_factor, method='relative', floor=.1, cap=2, asc=True)

        # score of the quality factor: biggest the best
        s2[t] = fs.fn_score(data[t], by=quality_factor, method='relative', floor=0, asc=False)

        # combining both price factor and quality factors picking 30 best companies from these two scores
        s[t] = fs.combine_score(s1[t], s2[t], n=top_k)
        signal[t] = list(s[t].index) # return buy vs sell signal 
    
    df = fs.backtest(signal=signal, data=data, m=3, cost=.001, rf_rate=.01)
    return df
```

```python
df_magic_formula_greenblatt = magic_formula_backtest(
    price_factor='ROC', 
    quality_factor='EV/EBIT', 
    factor_function=factor_as_greenblatt
    )
fs.draw_return(df_magic_formula_greenblatt)
```

### Test Youtube Managers's method: 2+ year ROE Avg & EBITDA

```python
# EV/EBIT + GP/A
terms = fs.set_terms(trade_start='2011Q1', trade_end='2021Q1')
data = {}
for t in terms:
    data[t] = fs.fn_consolidated(otp=MY_ENV_VAR, term=t)
clear_output()
print("data succesfully imported")
s = {}
s1 = {}
s2 = {}
signal = {}
for t in terms:
    prev_t = fs.quarters_before(terms, t, 4)

    data[t]["Market Cap"] = data[t]["Price"] * data[t]["Shares"]
    data[t]["Net Cash"] = data[t]["Cash & Equivalents"] - data[t]["Current Liabilities"]
    data[t]["EV/EBITDA"] = ( data[t]["Market Cap"] - data[t]["Net Cash"] ) / data[t]["EBITDA"]

    # quality factor
    data[t]["ROE"] = data[t]["Net Income"] / (data[t]['Total Assets'] - data[t]['Total Liabilities'])
    data[t]["Avg ROE"] =  ( data[t]['ROE'] + data[prev_t]['ROE'] ) / 2

    # price
    s1[t] = fs.fn_score(data[t], by='EV/EBITDA', method='relative', floor=.1, cap=2, asc=True)

    # quality
    s2[t] = fs.fn_score(data[t], by='Avg ROE', method='relative', floor=0, asc=False)

    # combine price factor and quality factor: pick best 30 companies
    s[t] = fs.combine_score(s1[t], s2[t], n=30)    
    signal[t] = list(s[t].index)

df_magic_formula_yt_manager = fs.backtest(signal=signal, data=data, m=3, cost=.001, rf_rate=.01)
fs.draw_return(df_magic_formula_yt_manager)
```

### When used 1 Year ROE

```python
# price factor
def factor_as_yt_manager(df):
    # price factor
    df["Market Cap"] = df["Price"] * df["Shares"]
    df["Net Cash"] = df["Cash & Equivalents"] - df["Current Liabilities"]
    df["EV/EBITDA"] = ( df["Market Cap"] - df["Net Cash"] ) / df["EBITDA"]

    # quality factor
    df["ROE"] = df["Net Income"] / (df['Total Assets'] - df['Total Liabilities'])
    return df
```

```python
# sky blue is NYSE and red is corresponding company

df_magic_formula_yt_manager = magic_formula_backtest(
    price_factor='ROE', 
    quality_factor='EV/EBITDA', 
    factor_function=factor_as_yt_manager
    )
fs.draw_return(df_magic_formula_yt_manager)
```

### Test Gray & Carlisle Method

```python
# EV/EBIT + GP/A
terms = fs.set_terms(trade_start='2011Q1', trade_end='2021Q1')
data = {}
for t in terms:
    data[t] = fs.fn_consolidated(otp=MY_ENV_VAR, term=t)
clear_output()
print("data succesfully imported")
s = {}
s1 = {}
s2 = {}
signal = {}
for t in terms:
    prev_t = fs.quarters_before(terms, t, 4)
    # price
    data[t]["Market Cap"] = data[t]["Price"] * data[t]["Shares"]
    data[t]["EV"] = data[t]["Market Cap"] + data[t]['Total Liabilities'] - data[t]["Cash & Equivalents"]
    data[t]["EV/EBIT"] = data[t]["EV"] / data[t]["EBIT"]
    
    # quality
    data[t]['Avg Assets'] = ( data[t]['Total Assets'] + data[prev_t]['Total Assets'] ) / 2
    data[t]['GP/A'] = data[t]['Gross Profit'] / data[t]['Avg Assets']
    
    # price
    s1[t] = fs.fn_score(data[t], by='EV/EBIT', method='relative', floor=.1, cap=2, asc=True)

    # quality
    s2[t] = fs.fn_score(data[t], by='GP/A', method='relative', floor=0, asc=False)

    # combine price factor and quality factor: pick best 30 companies
    s[t] = fs.combine_score(s1[t], s2[t], n=30)    
    signal[t] = list(s[t].index)

df_gray_carlisle = fs.backtest(signal=signal, data=data, m=3, cost=.001, rf_rate=.01)
fs.draw_return(df_gray_carlisle)
```

### PER & GP/A Method

```python
# PER + GP/A
terms = fs.set_terms(trade_start='2011Q1', trade_end='2021Q1')
data = {}
for t in terms:
    data[t] = fs.fn_consolidated(otp=MY_ENV_VAR, term=t)
clear_output()
print("data succesfully imported")

s = {}
s1 = {}
s2 = {}
signal = {}
for t in terms:
    prev_t = fs.quarters_before(terms, t, 4)
    data[t]['PER'] = data[t]['Price_M3'] / data[t]['EPS']
    data[t]['Avg Assets'] = ( data[t]['Total Assets'] + data[prev_t]['Total Assets'] ) / 2
    
    data[t]['GP/A'] = data[t]['Gross Profit'] / data[t]['Avg Assets']
    
    # price
    s1[t] = fs.fn_score(data[t], by='PER', method='relative', floor=.1, cap=2, asc=True)

    # quality
    s2[t] = fs.fn_score(data[t], by='GP/A', method='relative', floor=0, asc=False)

    # combine price factor and quality factor: pick best 30 companies
    s[t] = fs.combine_score(s1[t], s2[t], n=30)    
    signal[t] = list(s[t].index)
df_per_gpa = fs.backtest(signal=signal, data=data, m=3, cost=.001, rf_rate=.01)

# PER & GP/A
fs.draw_return(df_per_gpa) # NASDAQ companies backtest
```
